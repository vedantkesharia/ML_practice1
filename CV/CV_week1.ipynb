The arrangement of pixels in RGB (color) images and grayscale images differs in terms of the number of channels and the information each channel represents:

RGB (Color) Image:

In an RGB image, each pixel is represented by three values: Red (R), Green (G), and Blue (B).
The pixel is essentially a point in a 3D color space, where the combination of R, G, and B values determines the color of that pixel.
RGB images have three channels, and each channel stores color information.
The order of channels is typically (R, G, B), where each channel contains intensity values ranging from 0 to 255 for each color component.

Grayscale Image:

In a grayscale image, each pixel is represented by a single intensity value.
Grayscale images have only one channel, representing the intensity of the pixel.
The intensity value typically ranges from 0 (black) to 255 (white), with intermediate values representing shades of gray.
The grayscale image does not contain color information; it represents only the brightness or intensity of the image at each pixel location.
In summary, the key difference is that RGB images have three color channels (R, G, B) and represent color, while grayscale images have a single intensity channel and represent only the brightness or grayscale information. The arrangement of pixel values in RGB images and grayscale images is fundamentally different due to the number of channels and the type of information stored in each channel.

# Filter/Kernel/Convolution Operations

Convolution is a mathematical operation often used in image processing and computer vision. It involves applying a kernel (also known as a filter or mask) to an input image to perform operations like blurring, sharpening, edge detection, and more. The kernel is a small matrix, typically a square matrix, that slides over the input image, and at each position, it computes a weighted sum of the pixel values in the region covered by the kernel. This weighted sum forms the output pixel value in the resulting image.

1. Defining a kernel operation - https://www.youtube.com/watch?v=Etksi-F5ug8
  - watch the above video and understand what a convolution operation is
  - Write a function that applies any kernel convolution to any image input (pass kernel and image as arguments).
  - assume that the image is grayscale so there is only one channel instead of the usual RBG channels
  - note that kernels will always be a square matrix while an image can be of any shape
  - hint : use slicing to prevent for loops while performing addition operations

```python
from scipy.signal import convolve2d
import numpy as np

def convolution(kernel, image):
    image_height, image_width = image.shape
    kernel_height, kernel_width = kernel.shape
    
    padding_height = kernel_height // 2
    padding_width = kernel_width // 2
    
    result = np.zeros((image_height, image_width))
    
    padded_image = np.pad(image, ((padding_height, padding_height), (padding_width, padding_width)), mode='constant')
    
    for i in range(image_height):
        for j in range(image_width):
            result[i, j] = np.sum(padded_image[i:i+kernel_height, j:j+kernel_width] * kernel)
    
    return result
```

Padding is an essential concept in Convolutional Neural Networks (CNNs) used in computer vision tasks. It refers to the process of adding extra, usually zero-valued, pixels around the borders of an image before applying convolutional operations. Padding serves several purposes in CNNs:

Preserving Spatial Dimensions: Convolutional layers without padding reduce the spatial dimensions of the input feature map. Padding helps maintain the spatial dimensions of the feature maps, which can be crucial in many cases, especially when you want to preserve information at the borders of the input image.

Avoiding Information Loss: Without padding, the convolutional operation discards some information from the edges of the input image, making it difficult for the network to learn features at the image boundaries.

Centering the Convolution: Padding ensures that the convolution operation is centered at each pixel of the input image. This centering is important for accurately detecting features at different locations in the image.

In Convolutional Neural Networks (CNNs), the term "stride" refers to a hyperparameter that determines the step size at which the convolutional filter (kernel) moves across the input data or feature map. The stride controls how much the filter shifts (slides) over the input data when performing the convolution operation.
