**Regression Metrics**
1. MAE
Mean Absolute Error (MAE) is a regression metric that measures the average absolute difference between the predicted values and the actual values. It is calculated as follows:
MAE = Σ |y_true - y_pred| / n

where:
y_true is the actual value
y_pred is the predicted value
n is the number of observations
MAE is a simple metric that is easy to understand and interpret. It is also relatively insensitive to outliers, which makes it a good choice for regression problems with noisy data. However, MAE does not take into account the direction of the errors. For example, if a model consistently underestimates the actual values by 1 unit, the MAE will be the same as if it consistently overestimates the actual values by 1 unit.

2.MSE

Mean Squared Error (MSE) is a regression metric that measures the average squared difference between the predicted values and the actual values. It is calculated as follows:

MSE = Σ (y_true - y_pred)^2 / n

where:
y_true is the actual value
y_pred is the predicted value
n is the number of observations
MSE is a more sensitive metric than MAE, because it takes into account the direction of the errors. This means that a model with a higher MSE is considered to be worse than a model with a lower MSE. However, MSE is also more sensitive to outliers. This means that if there are a few outliers in the data, they can have a disproportionately large impact on the MSE.

Advantages:
MSE is a more sensitive metric than MAE, which means that it can be used to distinguish between good and bad models more accurately.
MSE is also a more intuitive metric, as it is based on the squared error between the predicted and actual values.

Disadvantages:
MSE is more sensitive to outliers than MAE, which means that a few outliers can have a disproportionately large impact on the MSE.
MSE can be difficult to interpret when the data is not normally distributed.


3.RMSE
Root Mean Squared Error (RMSE) is a regression metric that measures the average squared difference between the predicted values and the actual values. It is calculated as follows:

RMSE = √(Σ (y_true - y_pred)^2 / n)

where:
y_true is the actual value
y_pred is the predicted value
n is the number of observations
RMSE is similar to MSE, but it takes the square root of the MSE, which makes it easier to interpret. RMSE is also less sensitive to outliers than MSE. The best regression metric to use depends on the specific problem you are trying to solve. If you are looking for a metric that is sensitive to the direction of the errors and is not too sensitive to outliers, then RMSE is a good choice.

Advantages:
RMSE is less sensitive to outliers than MSE, which means that a few outliers will not have a disproportionately large impact on the RMSE.
RMSE is also easier to interpret than MSE, as it is expressed in the same units as the target variable.

Disadvantages:
RMSE is not as sensitive to the direction of the errors as MSE, which means that it may not be as useful for distinguishing between good and bad models.
RMSE can be difficult to interpret when the data is not normally distributed.

4. R2 score
R-squared, also known as the coefficient of determination, is a statistical measure of how well the regression line fits the data. It is calculated as follows:

R^2 = 1 - (Σ (y_true - y_pred)^2 / Σ (y_true - y_mean)^2)

where:
y_true is the actual value
y_pred is the predicted value
y_mean is the mean of the actual values
n is the number of observations
R-squared can be interpreted as the percentage of the variance in the target variable that is explained by the regression model. A higher R-squared value indicates a better fit of the model to the data.


However, R-squared can be misleading if the data is not normally distributed. It can also be misleading if there are outliers in the data.
Here are some of the advantages and disadvantages of R-squared:
Advantages:
R-squared is a simple metric that is easy to understand and interpret.
R-squared is a good measure of how well the regression line fits the data.

Disadvantages:
R-squared can be misleading if the data is not normally distributed.
R-squared can be misleading if there are outliers in the data.

5. Adjusted R2 Score
Adjusted R-squared is a modification of R-squared that takes into account the number of predictors in the model. It is calculated as follows:

Adjusted R^2 = 1 - (n - 1) * (1 - R^2) / n - k

where:
n is the number of observations
k is the number of predictors
R^2 is the coefficient of determination
Adjusted R-squared is a more conservative measure of fit than R-squared, because it penalizes the model for adding additional predictors. This is because adding additional predictors can increase the R-squared value, even if the new predictors do not actually improve the fit of the model.
Advantages:
Adjusted R-squared is a more conservative measure of fit than R-squared.
Adjusted R-squared is less likely to be misleading if the data is not normally distributed.
Adjusted R-squared is less likely to be misleading if there are outliers in the data.

Disadvantages:
Adjusted R-squared is a more complicated metric than R-squared.
Adjusted R-squared is not as easy to interpret as R-squared.


The point of regression metrics is to minimize error in different regression algorithms. Regression metrics are used to evaluate the performance of a regression model. They measure the difference between the predicted values and the actual values.

### **Regression Algorithms**

### 1. Linear Regression
Linear regression is a statistical method that uses a linear model to predict the value of a dependent variable from one or more independent variables. The linear model is a mathematical equation that describes the relationship between the independent and dependent variables.The dependent variable is the variable that is being predicted, and the independent variables are the variables that are used to predict the dependent variable. The linear model is typically of the form:

y = ax + b

where:
y is the dependent variable
x is the independent variable
a is the slope of the line
b is the y-intercept
The slope of the line tells us how much the dependent variable changes when the independent variable changes by one unit. The y-intercept tells us the value of the dependent variable when the independent variable is zero.

Linear regression can be used to solve a variety of problems, such as:

Predicting the price of a house based on its size and location
Predicting the number of sales based on the amount of advertising
Predicting the risk of a patient developing a disease based on their age and health history
Linear regression is a powerful tool that can be used to make predictions about the real world. However, it is important to note that linear regression is not always the best solution. In some cases, other statistical methods, such as logistic regression or nonlinear regression, may be more appropriate.

Here are some of the advantages and disadvantages of linear regression:

Advantages:
Linear regression is relatively simple to understand and implement.
Linear regression can be used to solve a variety of problems.
Linear regression is relatively efficient, meaning that it can be used to analyze large datasets.

Disadvantages:
Linear regression is not always the best solution.
Linear regression can be sensitive to outliers.
Linear regression can be biased if the independent variables are not normally distributed.
Overall, linear regression is a powerful tool that can be used to make predictions about the real world. However, it is important to be aware of its limitations before using it.

### 2. Logistic Regression
Logistic regression is a statistical method that is used to predict the probability of a categorical dependent variable. The categorical dependent variable can have two or more possible values, such as "yes" or "no", "true" or "false", or "spam" or "not spam". Logistic regression is a type of regression analysis, but it is different from linear regression in that it is used to predict a categorical variable, rather than a continuous variable.

The logistic regression model is a mathematical equation that describes the relationship between the independent variables and the probability of the dependent variable taking on a particular value. The model is typically of the form:

log(P(y = 1)) = ax + b

where:
P(y = 1) is the probability of the dependent variable taking on the value "1"
x is the independent variable
a is the slope of the line
b is the y-intercept
The slope of the line tells us how much the probability of the dependent variable taking on the value "1" changes when the independent variable changes by one unit. The y-intercept tells us the probability of the dependent variable taking on the value "1" when the independent variable is zero.

Logistic regression can be used to solve a variety of problems, such as:

Predicting whether a patient will develop a disease
Predicting whether a customer will click on an ad
Predicting whether an email is spam
Logistic regression is a powerful tool that can be used to make predictions about the real world. However, it is important to note that logistic regression is not always the best solution. In some cases, other statistical methods, such as linear regression or decision trees, may be more appropriate.

Here are some of the advantages and disadvantages of logistic regression:

Advantages:
Logistic regression can be used to predict a categorical variable.
Logistic regression is relatively simple to understand and implement.
Logistic regression can be used to solve a variety of problems.

Disadvantages:
Logistic regression can be sensitive to outliers.
Logistic regression can be biased if the independent variables are not normally distributed.
Logistic regression can be computationally expensive, especially for large datasets.
Overall, logistic regression is a powerful tool that can be used to make predictions about the real world. However, it is important to be aware of its limitations before using it.

Cost Function
A cost function is a function that measures the error between the predicted values and the actual values in a machine learning model. It is used to evaluate the performance of the model and to optimize the model parameters.
Cost function in terms of linear regression: 

MSE = (1/2m) * Σ(predicted_i - actual_i)^2

m is the number of data points.
predicted_i is the predicted value for the i-th data point using the linear equation.
actual_i is the actual value for the i-th data point.
The linear equation is typically expressed as: predicted_i = slope * x_i + bias, where slope is the coefficient for the feature x_i, and bias is the intercept term.

A confusion matrix is typically a square matrix that has rows and columns representing the actual and predicted classes. It contains four important values:

True Positives (TP): The number of instances that were correctly predicted as positive (i.e., the model predicted the positive class, and it was correct).

True Negatives (TN): The number of instances that were correctly predicted as negative (i.e., the model predicted the negative class, and it was correct).

False Positives (FP): The number of instances that were incorrectly predicted as positive (i.e., the model predicted the positive class, but it was incorrect).

False Negatives (FN): The number of instances that were incorrectly predicted as negative (i.e., the model predicted the negative class, but it was incorrect).

### Choosing optimal 'k'

**Above, we chose n_neighbours to be equal to 1. Choosing a small value of K leads to unstable decision boundaries. <br>
We need to select n_neighbours by calculating the accuracy for every value of n from 1 to 60 and then choosing the one which gives the highest accuracy.**

```python
# Do the same as we did above, but this time make a loop from n = 1 to n = 60 and append the accuracy
# for each in a list
# Your code starts here. 
accuracy_rate = []

for i in range(1,61):
    knn = KNeighborsClassifier(n_neighbors=i)
    score = cross_val_score(knn,df_feat,df['TARGET CLASS'],cv=10)
    accuracy_rate.append(score.mean())
print(accuracy_rate)
# Your code ends here
```

**Plot a graph of K value vs Accuracy**

```python
# Your code starts here. 

# Plot K value vs. Accuracy
plt.figure(figsize=(10, 6))
plt.plot(range(1, 61), accuracy_rate, marker='o', linestyle='dashed')
plt.title('Accuracy vs K Value')
plt.xlabel('K Value')
plt.ylabel('Accuracy')
plt.show()
# Your code ends here
```

**Choose the best value of n_neighbours and give a reason why and also print the accuracy**

```python
# Your code starts here. 
best_n_neighbors_index = np.argmax(accuracy_rate)

# Determine the best n_neighbors value based on the index
best_n_neighbors = n_neighbors_range[best_n_neighbors_index]

# Print the best n_neighbors and its corresponding accuracy score
print(f"Best n_neighbors: {best_n_neighbors}")
print(f"Highest Accuracy: {accuracy_rate[best_n_neighbors_index]:.3f}")
# Your code ends here
```
