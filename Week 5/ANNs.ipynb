**What is Artificial Neural Network?**

Artificial Neural Network ANN is an efficient computing system whose central theme is borrowed from the analogy of biological neural networks. ANNs are also named as “artificial neural systems,” or “parallel distributed processing systems,” or “connectionist systems.” ANN acquires a large collection of units that are interconnected in some pattern to allow communication between the units. These units, also referred to as nodes or neurons, are simple processors which operate in parallel.

Every neuron is connected with other neuron through a connection link. Each connection link is associated with a weight that has information about the input signal. This is the most useful information for neurons to solve a particular problem because the weight usually excites or inhibits the signal that is being communicated. Each neuron has an internal state, which is called an activation signal. Output signals, which are produced after combining the input signals and activation rule, may be sent to other units.

The Neural Network Architecture

Before jumping into the architecture, let’s take a look at some of the components of a Neural Network.

**The Input Layer** — Represents the input variables plus the bias term. Hence if there are n input variables, the size of the input layer is n + 1, where + 1 is the bias term

**The Hidden Layer/ Layers —** These signify neurons where all mathematical calculations are done. Note a given neural network can have more than one neuron in a hidden layer or multiple hidden layers as well

**The Activation Function** — Converts the output of a given layer before passing on the information to consecutive layers. Activation functions are mathematical equations that determine the output of a given neural network. The is a part of each neuron in the hidden layers and determines output relevant for prediction

**The Output Layer** — The final “output prediction” of the network

**Forward Propagation** — Calculating the output of each iteration from the input layer to the output layer

**Backward Propagation** — Calculates revised weights (w1, w2, w3, and b1) after each forward propagation by analyzing the derivative of the cost function used to optimize the model output

**Learning Rate —** Determined the percentage change attributed to each weight and bias term after every backward propagation, i.e. controls the speed at which the model learns information about the data

## **STEPS ON HOW A NEURAL NETWORK MODEL IS TRAINED**

Backpropagation is a fundamental algorithm used to train neural networks by computing gradients of the loss function with respect to the model's weights. Let's break down the key concepts and steps involved in backpropagation using the gradient notation ∂(loss)/∂(weight):

1. **Forward Pass**:
   - During the forward pass, input data is fed through the neural network to compute predictions. These predictions are compared to the actual target values, and a loss function (e.g., mean squared error or cross-entropy loss) is computed to quantify how well the model is performing.

2. **Backward Pass (Backpropagation)**:
   - The backward pass is where the gradients of the loss with respect to each weight parameter are calculated. The goal is to determine how a small change in each weight would affect the loss.

   - Start with the output layer and compute the gradient of the loss with respect to the output of that layer (∂(loss)/∂(output)). This gradient depends on the specific loss function used (e.g., ∂(loss)/∂(output) = ∂(loss_function)/∂(output)).
   
   - Then, for each layer going backward through the network, calculate the gradient of the layer's output with respect to the weights (∂(output)/∂(weight)). This is typically done by applying the chain rule.

   - Multiply the gradients obtained in the previous two steps to get the gradient of the loss with respect to the weights in that layer (∂(loss)/∂(weight) = (∂(loss)/∂(output)) * (∂(output)/∂(weight))).

   - This process is repeated for each weight and each layer, propagating the gradients backward through the network.

3. **Gradient Descent Update**:
   - After computing the gradients of the loss with respect to the weights for all layers, these gradients are used to update the weights. The update rule typically follows gradient-based optimization algorithms like gradient descent:
   
     `weight = weight - learning_rate * ∂(loss)/∂(weight)`

   - The learning rate controls the step size in the weight updates. By subtracting the gradient of the loss with respect to the weight from the weight itself, the weights are adjusted to minimize the loss function.

4. **Iterative Process**:
   - Steps 1 to 3 are repeated iteratively for a fixed number of iterations (epochs) or until convergence is achieved. During each iteration, the weights are updated to improve the model's performance on the training data, gradually reducing the loss.

In summary, backpropagation is the process of computing gradients of the loss function with respect to the model's weights by propagating errors backward through the network. These gradients are then used to adjust the weights to minimize the loss, enabling the neural network to learn and make better predictions over time.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import math
import sklearn
```

```python
titanic_data = pd.read_csv("titanic.csv")
titanic_data.head()
```

Keras is a popular deep learning library used for building and training neural networks.

```python
#import all the prerequisites like tensorflow and keras
import keras
from keras import backend as K
from keras.models import Sequential
from keras.layers import Activation,Dense,Input
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
#create a model using sequential
#add dense layers and activation function
model = Sequential([
    Dense(16, input_shape=(9,), activation='relu'), 
    Dense(32, activation='relu'),
    Dense(2, activation='softmax')
])
```

model = Sequential([ ... ]): This line creates a new Keras Sequential model. A Sequential model is a linear stack of layers, where you can simply add one layer at a time, and they are executed sequentially from input to output.

Dense(16, input_shape=(9,), activation='relu'): This is the first layer of your neural network. It's a dense layer, also known as a fully connected layer, with 16 units. Here's what each argument means:

16: The number of neurons (units) in this layer.
input_shape=(1,): This specifies the shape of the input data. In this case, you have a one-dimensional input with a single feature, so the input shape is (1,).
activation='relu': The activation function used in this layer is Rectified Linear Unit (ReLU), which is a common activation function in neural networks.
Dense(32, activation='relu'): This is the second layer of your neural network. It's another dense layer with 32 units and uses the ReLU activation function. This layer is fully connected to the previous layer.

Dense(2, activation='softmax'): This is the output layer of your neural network. It's a dense layer with 2 units, and the activation function used is softmax. In many cases, softmax is used in the output layer for classification tasks with more than two classes. It produces a probability distribution over the possible class labels, making it suitable for multi-class classification problems.

So, in summary, your neural network architecture consists of three layers:

Input Layer: One input feature (1-dimensional) with 16 units and ReLU activation.
Hidden Layer: 32 units with ReLU activation.
Output Layer: 2 units with softmax activation for multi-class classification.
This architecture is suitable for tasks where you want to take a single input feature, pass it through a neural network, and make predictions in one of the two classes. Depending on your specific task and dataset, you may need to adjust the number of units, layers, and activation functions to achieve the desired performance.

```python
embark = pd.get_dummies(titanic_data['Embarked'],drop_first=True)
embark.head(5)
```

```python
titanic_data = pd.concat([titanic_data,sex,embark,Pcl],axis=1) #concating the new columns with dataset
titanic_data.head(5)
```

```python
X= titanic_data.drop('Survived',axis=1) #independent variable -> predictor 
# X.columns = X.columns.astype(str) 
y = titanic_data['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
model.compile(Adam(learning_rate=.0001),loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

model.compile: This is a method provided by Keras that is used to configure the training process of the model. It prepares the model for training by specifying various settings such as the optimizer, loss function, and evaluation metrics.

Adam(learning_rate=0.0001): In this part, you are specifying the optimizer to be used during training. Adam (short for Adaptive Moment Estimation) is an optimization algorithm that is widely used for training deep neural networks. The learning_rate parameter sets the step size or learning rate for the optimizer. In this case, it's set to 0.0001, which controls how quickly the model's weights are updated during training.

loss='sparse_categorical_crossentropy': This part specifies the loss function that will be used during training. The loss function measures the difference between the predicted values and the actual target values, and the model tries to minimize this loss. 'sparse_categorical_crossentropy' is a common choice for multi-class classification problems where the target labels are integers (e.g., 0, 1, 2) rather than one-hot encoded vectors.

metrics=['accuracy']: Here, you specify the evaluation metric(s) that you want to monitor during training. In this case, you are using 'accuracy', which calculates the accuracy of the model on the training data. Accuracy is a commonly used metric for classification tasks and measures the fraction of correctly classified samples.

```python
model.loss = 'sparse_categorical_crossentropy'
```

```python
model.loss
```

```python
model.fit(X_train,y_train,validation_split=0.20,batch_size=10,epochs=20,shuffle=True,verbose=2)
```

batch_size=10: This parameter determines the batch size used for training. In deep learning, training is often done in batches rather than on the entire dataset at once. A batch size of 10 means that the model will process 10 samples at a time, calculate the gradient, and update its weights. This process is repeated for all batches until one training epoch is complete.

epochs=20: The epochs parameter specifies the number of times the entire training dataset (all batches) will be processed by the model during training. In this case, the model will go through the entire dataset 20 times, updating its weights with each pass.

shuffle=True: Setting shuffle to True indicates that the training data will be shuffled before each epoch. Shuffling helps ensure that the model doesn't learn any sequence-related patterns in the data and generalizes better.

verbose=2: The verbose parameter controls the amount of information displayed during training. Setting it to 2 means that training progress updates will be printed to the console for each epoch. You can adjust this value to control the verbosity of the training process. For example, setting it to 1 would display progress bar without per-epoch details, and setting it to 0 would run training silently.

The learning rate controls the step size in the weight updates. By subtracting the gradient of the loss with respect to the weight from the weight itself, the weights are adjusted to minimize the loss function.

**Predictions**

```python
#use model.predict to predict the survival rate on the test dataset and print the predictions
#Try to understand what does these numbers in predict means
# Model.predict passes the input vector through the trained(fit) model and returns the output tensor for each datapoint.
# the numbers in predict are tensors which predict probability(0,1) of survival of that person. if its close to 0, person dies. if its close to 1, person survives.   

predictions = model.predict(X_test)
```

```python
print("Predictions:")
for i in predictions:
    print(i)
```

```python
threshold = 0.5  # we set the threshold to 0.5


for i, prediction in enumerate(predictions):
    probability_of_survival = prediction[1]  # Probability of surviving (class 1)
    
    if probability_of_survival > threshold:  #if > 0.5 person survived, if less than 0.5 did not 
        print(f"Sample {i + 1}: Predicted Probability of Survival: {probability_of_survival:.4f} - Person Predicted as Survived")
    else:
        print(f"Sample {i + 1}: Predicted Probability of Survival: {probability_of_survival:.4f} - Person Predicted as Not Survived")
```
